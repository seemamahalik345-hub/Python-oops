```python
#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.


from abc import ABC, abstractmethod
import math

# Abstract class Shape
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Derived class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Derived class Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Example usage:
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Area of Circle: {circle.area():.2f}")
print(f"Area of Rectangle: {rectangle.area()}")
```

```python
#3.  Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.


class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

    def get_type(self):
        return self.vehicle_type

class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def get_brand(self):
        return self.brand

class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery_capacity = battery_capacity

    def get_battery_info(self):
        return f"Battery Capacity: {self.battery_capacity} kWh"

# Example usage:
my_vehicle = Vehicle("Sedan")
print(f"My vehicle is a: {my_vehicle.get_type()}")

my_car = Car("SUV", "Toyota")
print(f"My car is a: {my_car.get_type()} from {my_car.get_brand()}")

my_electric_car = ElectricCar("Hatchback", "Tesla", 75)
print(f"My electric car is a: {my_electric_car.get_type()} from {my_electric_car.get_brand()} with {my_electric_car.get_battery_info()}")
```

```python
#4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.

class Bird:
    def fly(self):
        print("Birds generally fly.")

class Sparrow(Bird):
    def fly(self):
        print("Sparrows fly high in the sky.")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly; they swim.")

# Example usage:
bird = Bird()
bird.fly()

sparrow = Sparrow()
sparrow.fly()

penguin = Penguin()
penguin.fly()

def make_it_fly(bird_obj):
    bird_obj.fly()

print("\n--- Using a common function ---")
make_it_fly(bird)
make_it_fly(sparrow)
make_it_fly(penguin)
```

```python
#5.Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, initial_balance=0):
        # Private attribute convention using a single underscore
        self._balance = initial_balance
        print(f"Account created with initial balance: ${self._balance:.2f}")

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self._balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if self._balance >= amount:
                self._balance -= amount
                print(f"Withdrew ${amount:.2f}. New balance: ${self._balance:.2f}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def check_balance(self):
        print(f"Current balance: ${self._balance:.2f}")
        return self._balance

# Example usage:
my_account = BankAccount(100)
my_account.check_balance()

my_account.deposit(50)
my_account.withdraw(30)
my_account.check_balance()

my_account.withdraw(200) # Attempt to overdraw
my_account.deposit(-10) # Attempt invalid deposit
my_account.check_balance()
```

### 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

```python
class Instrument:
    def play(self):
        print("Playing a generic instrument sound.")

class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar strings.")

class Piano(Instrument):
    def play(self):
        print("Pressing the piano keys.")

# Function demonstrating polymorphism
def make_it_play(instrument_obj):
    instrument_obj.play()

# Example usage:
instrument = Instrument()
guitar = Guitar()
piano = Piano()

print("--- Individual Instrument Sounds ---")
instrument.play()
guitar.play()
piano.play()

print("\n--- Demonstrating Runtime Polymorphism ---")
make_it_play(instrument)
make_it_play(guitar)
make_it_play(piano)
```

### 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

```python
class MathOperations:
    @classmethod
    def add_numbers(cls, num1, num2):
        print(f"Using class method: Adding {num1} and {num2}")
        return num1 + num2

    @staticmethod
    def subtract_numbers(num1, num2):
        print(f"Using static method: Subtracting {num2} from {num1}")
        return num1 - num2

# Example usage:

# Calling the class method
sum_result = MathOperations.add_numbers(10, 5)
print(f"Sum: {sum_result}")

# Calling the static method
diff_result = MathOperations.subtract_numbers(20, 7)
print(f"Difference: {diff_result}")

# You can also call them via an instance, though it's less common for these types of methods
math_obj = MathOperations()
instance_sum = math_obj.add_numbers(8, 2)
print(f"Instance Sum: {instance_sum}")

instance_diff = math_obj.subtract_numbers(15, 3)
print(f"Instance Difference: {instance_diff}")
```

### 8. Implement a class Person with a class method to count the total number of persons created.

```python
class Person:
    _total_persons = 0  # Class variable to keep track of total persons

    def __init__(self, name):
        self.name = name
        Person._total_persons += 1  # Increment the count each time a new Person object is created
        print(f"Person '{self.name}' created. Total persons: {Person._total_persons}")

    @classmethod
    def get_total_persons(cls):
        """Class method to return the total number of Person objects created."""
        return cls._total_persons

# Example usage:
p1 = Person("Alice")
p2 = Person("Bob")
p3 = Person("Charlie")

print(f"\nFinal total number of persons created: {Person.get_total_persons()}")

# You can also call the class method via an instance, though it's less common for this purpose
print(f"Total persons (via instance): {p1.get_total_persons()}")
```

### 9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".



```python
class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Example usage:
f1 = Fraction(3, 4)
f2 = Fraction(1, 2)
f3 = Fraction(7, 1)

print(f"Fraction 1: {f1}")
print(f"Fraction 2: {f2}")
print(f"Fraction 3: {f3}")

# You can also try to create an invalid fraction
try:
    f4 = Fraction(5, 0)
except ValueError as e:
    print(f"Error creating fraction: {e}")
```

### 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Can only add a Vector object to another Vector object")

# Example usage:
v1 = Vector(2, 3)
v2 = Vector(5, 7)

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")

v3 = v1 + v2
print(f"Vector 1 + Vector 2: {v3}")

# Attempt to add with a non-Vector object (will raise an error)
try:
    v4 = v1 + 10
except TypeError as e:
    print(f"Error: {e}")
```

### 12. ` Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

```python
class Student:
    def __init__(self, name, grades):
        self.name = name
        # Ensure grades is a list of numbers
        if not all(isinstance(grade, (int, float)) for grade in grades):
            raise ValueError("Grades must be a list of numbers.")
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0.0 # Return 0 if there are no grades to avoid division by zero
        return sum(self.grades) / len(self.grades)

# Example usage:
student1 = Student("Alice", [90, 85, 92, 88])
print(f"{student1.name}'s average grade: {student1.average_grade():.2f}")

student2 = Student("Bob", [75, 80, 70])
print(f"{student2.name}'s average grade: {student2.average_grade():.2f}")

student3 = Student("Charlie", []) # Student with no grades
print(f"{student3.name}'s average grade: {student3.average_grade():.2f}")

try:
    student4 = Student("David", [100, "A"]) # Invalid grades
except ValueError as e:
    print(f"Error creating student: {e}")
```

### 13.  Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.

```python
class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        if length < 0 or width < 0:
            raise ValueError("Dimensions cannot be negative.")
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Example usage:
rectangle1 = Rectangle()
rectangle1.set_dimensions(10, 5)
print(f"Rectangle 1 - Length: {rectangle1.length}, Width: {rectangle1.width}, Area: {rectangle1.area()}")

rectangle2 = Rectangle()
rectangle2.set_dimensions(7, 3)
print(f"Rectangle 2 - Length: {rectangle2.length}, Width: {rectangle2.width}, Area: {rectangle2.area()}")

try:
    rectangle3 = Rectangle()
    rectangle3.set_dimensions(-2, 4) # Attempt with negative dimension
except ValueError as e:
    print(f"Error setting dimensions for Rectangle 3: {e}")
```
